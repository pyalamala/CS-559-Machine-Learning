```python
import math
import numpy as np
import pandas as pd
```

```python
X = np.array([[5.9, 4.6, 6.2, 4.7, 5.5, 5.0, 4.9, 6.7, 5.1, 6.0],
             [3.2, 2.9, 2.8, 3.2, 4.2, 3.0, 3.1, 3.1, 3.8, 3.0]])
```

```python
red = np.array([6.2,3.2])
green = np.array([6.6, 3.7])
blue = np.array([6.5, 3.0])
```

```python
def kmeans(X, red, green, blue):
    old_red = np.array([0,0])
    old_green = np.array([0,0])
    old_blue = np.array([0,0])
    converge = False
    i = 0
    while converge != True:
        i += 1
        red, green, blue = assign_points(X, red, green, blue)
        if ((red - old_red).all() and (green - old_green).all() and (blue - old_blue).all()):
            converge == True
        else:
            old_red = red
            old_green = green
            old_blue = blue
        print(f"Iteration {i} red: {red}")
        print(f"Iteration {i} green: {green}")
        print(f"Iteration {i} blue: {blue}")
    return red, green, blue
```

```python
def assign_points(X, red, green, blue):
    red_points = [[], []]
    green_points = [[], []]
    blue_points = [[], []]
    
    for i in range(0, len(X[0])):
#         print(f"current point: ({X[0][i]}, {X[1][i]})\n")
#         print(red[0])
#         print(red[1])
        red_distance = math.sqrt((X[0][i] - red[0])**2 + (X[1][i] - red[1])**2)
        green_distance = math.sqrt((X[0][i] - green[0])**2 + (X[1][i] - green[1])**2)
        blue_distance = math.sqrt((X[0][i] - blue[0])**2 + (X[1][i] - blue[1])**2)
        closest = min(red_distance, green_distance, blue_distance)
#         print(f"Red:{red_distance}\nGreen:{green_distance}\nBlue:{blue_distance}")
#         print(f"min_distance:{closest}")
        if(closest == red_distance):
            red_points[0].append(X[0][i])
            red_points[1].append(X[1][i])
        elif(closest == green_distance):
            green_points[0].append(X[0][i])
            green_points[1].append(X[1][i])
        elif(closest == blue_distance):
            blue_points[0].append(X[0][i])
            blue_points[1].append(X[1][i])
    red_points = np.array(red_points)
    green_points = np.array(green_points)
    blue_points = np.array(blue_points)
    if(red_points.size == []):
        new_red = red
    else:
        new_red = [np.mean(red_points[0]), np.mean(red_points[1])]
    if(green_points.size == 0):
        new_green = green
    else:
        new_green = [np.mean(green_points[0]), np.mean(green_points[1])]
    if(blue_points.size == 0):
        new_blue = blue
    else:
        new_blue = [np.mean(blue_points[0]), np.mean(blue_points[1])]
    
    
    return new_red, new_green, new_blue
```

```python
final_red, final_green, final_blue = kmeans(X, red, green, blue)
```
